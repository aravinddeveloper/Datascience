```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import PolynomialFeatures
import warnings
warnings.filterwarnings("ignore")
np.random.seed(42)
```

```python
loan = pd.read_csv("D:\Data Science\Machine Learning\Random Forest\Assignment\loan_data.csv")
loan.head()
```

```python
y=loan.pop('not.fully.paid')
```

```python
dummies = pd.get_dummies(loan['purpose'])
dummies.head()
```

```python
loan = pd.concat([loan,dummies],axis=1)
```

```python
loan.pop('purpose')
loan.head()
```

```python
train,test,train_y,test_y = train_test_split(loan,y,test_size=0.2,random_state=101)
```

```python
lr = LogisticRegression()
lr.fit(train,train_y)
y_pred = lr.predict(test)
print(accuracy_score(test_y,y_pred))
```

```python
dt = DecisionTreeClassifier()
dt.fit(train,train_y)
y_pred = dt.predict(test)
print(accuracy_score(test_y,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gs = GridSearchCV(dt,params,verbose=3)
```
